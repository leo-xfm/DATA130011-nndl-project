```python
import pickle
import os
```

```python
saved_models_dir = "saved_models"
files = os.listdir(saved_models_dir)
```

```python
train_loss = []
train_score = []
eval_score = []
# test_score = []
```

```python
for file in files:
    with open(f"saved_models/{file}/training_data.pkl", 'rb') as f:
        data = pickle.load(f)
    train_loss.append(data['train_loss'])
    train_score.append(data['train_score'])
    eval_score.append(data['eval_score_ls'])
    # test_score.append(data['test_score'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 3, figsize=(24, 20))

for idx, file in enumerate(files):
    ax[0].plot(train_loss[idx][:500], label=f'Train Loss {file}', linewidth=1)
    ax[1].plot(train_score[idx][:500], label=f'Train Score {file}', linewidth=1)
    ax[2].plot(eval_score[idx][:500], label=f'Eval Score {file}', linewidth=1)
    ax[0].set_title('Train Loss')
    ax[1].set_title('Train Score')
    ax[2].set_title('Eval Score')

for i in range(3):
    ax[i].set_xlabel('Epochs')
    ax[i].set_xlabel('Score')
    ax[i].legend()
    
plt.grid()
plt.show()
```
